```
def information_gain(left_tree,right_tree,current_impurity):

  total_len = len(left_tree)+len(right_tree)
  if total_len > 0 :
    information_gain = current_impurity - (len(left_tree)/total_len)*gini_impurity(left_tree) + (len(right_tree)/total_len)*gini_impurity(right_tree) 
    return information_gain
  else:
    return 0
```

```
def split_data(df,feature,value):
  print(type(value))
  left_tree = []
  right_tree = []
  if isinstance(value,np.int64):
    for index,k in enumerate(df[feature].values):
      if value <= k :
        left_tree.append(df.values[index])
      else:
        right_tree.append(df.values[index])
  elif isinstance(value, str):
    for index,k in enumerate(df[feature].values):
      if value == k :
        left_tree.append(df.values[index])
      else:
        right_tree.append(df.values[index])

  return left_tree,right_tree,value
```
